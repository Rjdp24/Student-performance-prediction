# Feature Engineering of the dataset

```python
import pandas as pd

df = pd.read_csv("data/student_performance_dataset.csv")

df['stress_sleep_ratio'] = df['stress_level'] / (df['sleep_hours'] + 1e-5)
df['motivation_stress_diff'] = df['motivation_score'] - df['stress_level']
df['screen_sleep_balance'] = df['screen_time'] / (df['sleep_hours'] + 1e-5)
df['rest_quality_score'] = df['sleep_quality'] * df['sleep_hours']
df['mental_health_index'] = (df['motivation_score'] + df['social_support']) - df['stress_level']
df['study_support_score'] = df['study_environment_rating'] + df['parental_involvement'] + df['physical_activity_days']
df['time_drain_score'] = df['commute_time'] + (df['part_time_job'] * 5)
df['room_disturbance_score'] = df['room_sharing'] * (6 - df['study_environment_rating'])

# df.to_csv("data/student_performance_dataset_with_engineered_features.csv", index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
numeric_df = df.select_dtypes(include=[np.number])
corr = numeric_df.corr()

plt.figure(figsize=(20, 18))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
